```python
import os
import cv2
import numpy as np
from tensorflow.keras.models import load_model

MODEL_PATH = "my_model.keras"
SAVE_DIR = "collected_samples"
os.makedirs(SAVE_DIR, exist_ok=True)

# Load model
model = load_model(MODEL_PATH)
print("Model loaded. Input shape:", model.input_shape)

# Helper: center/pad to 28x28 (MNIST-style)
def center_pad_to_28(img_bin):
    ys, xs = np.where(img_bin > 0)
    if len(xs) == 0:
        return None
    x_min, x_max = xs.min(), xs.max()
    y_min, y_max = ys.min(), ys.max()
    crop = img_bin[y_min:y_max+1, x_min:x_max+1]

    # Resize longest side to 20 (MNIST convention)
    h, w = crop.shape
    scale = 20.0 / max(h, w)
    new_w = max(1, int(round(w * scale)))
    new_h = max(1, int(round(h * scale)))
    resized = cv2.resize(crop, (new_w, new_h), interpolation=cv2.INTER_AREA)

    # Place centered on 28x28 canvas
    canvas = np.zeros((28, 28), dtype=np.uint8)
    y_off = (28 - new_h) // 2
    x_off = (28 - new_w) // 2
    canvas[y_off:y_off+new_h, x_off:x_off+new_w] = resized

    # Center by centroid (shift)
    M = cv2.moments(canvas)
    if M["m00"] != 0:
        cx = M["m10"] / M["m00"]
        cy = M["m01"] / M["m00"]
        shift_x = int(round(14 - cx))
        shift_y = int(round(14 - cy))
        T = np.float32([[1, 0, shift_x], [0, 1, shift_y]])
        canvas = cv2.warpAffine(canvas, T, (28, 28))

    return canvas

# Preprocess & predict for a single ROI (tries both polarities)
def preprocess_and_predict(roi_bgr):
    gray = cv2.cvtColor(roi_bgr, cv2.COLOR_BGR2GRAY)
    gray = cv2.GaussianBlur(gray, (5,5), 0)

    # adaptive threshold helps with variable lighting
    thr = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                cv2.THRESH_BINARY, 11, 2)
    thr_inv = cv2.bitwise_not(thr)  # inverted polarity alternative

    candidates = []
    for mode, bin_img in (("black_on_white", thr), ("white_on_black", thr_inv)):
        cnts, _ = cv2.findContours(bin_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        if not cnts:
            continue
        # pick largest contour (assume digit)
        c = max(cnts, key=cv2.contourArea)
        if cv2.contourArea(c) < 100:   # skip tiny noise
            continue
        mask = np.zeros_like(bin_img)
        cv2.drawContours(mask, [c], -1, 255, thickness=cv2.FILLED)
        digit_only = cv2.bitwise_and(bin_img, mask)

        img28 = center_pad_to_28(digit_only)
        if img28 is None:
            continue

        # Check if model has Rescaling/Normalization layer to avoid double-normalizing
        has_internal_rescale = any("Rescaling" in layer.__class__.__name__ or
                                   "Normalization" in layer.__class__.__name__ for layer in model.layers)
        if has_internal_rescale:
            x = img28.astype("float32").reshape(1,28,28,1)  # model rescales itself
        else:
            x = (img28.astype("float32") / 255.0).reshape(1,28,28,1)

        preds = model.predict(x, verbose=0)[0]
        top3_idx = preds.argsort()[-3:][::-1]
        top3 = [(int(i), float(preds[i])) for i in top3_idx]
        candidates.append((mode, img28, x, preds, top3))

    if not candidates:
        return None
    # choose highest-confidence candidate
    best = max(candidates, key=lambda c: c[3].max())
    mode, img28, x_input, preds, top3 = best
    pred = int(np.argmax(preds))
    conf = float(np.max(preds))
    return {"mode":mode, "pred":pred, "conf":conf, "top3":top3, "img28":img28, "raw":preds}

# Open camera
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise RuntimeError("Cannot open camera")

print("Press 's' to scan a frame and predict (press key 0-9 to save sample). 'q' to quit.")

current_processed = None  # store last processed 28x28

while True:
    ret, frame = cap.read()
    if not ret:
        break

    h, w, _ = frame.shape
    size = min(h, w, 360)
    x1 = w//2 - size//2
    y1 = h//2 - size//2
    x2 = x1 + size
    y2 = y1 + size
    cv2.rectangle(frame, (x1,y1), (x2,y2), (255,0,0), 2)
    cv2.putText(frame, "Place ONE digit inside box. Press 's' to scan.", (10,30),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,255,255), 2)

    cv2.imshow("Live", frame)
    key = cv2.waitKey(1) & 0xFF

    if key == ord('s'):   # capture + predict
        roi = frame[y1:y2, x1:x2]
        info = preprocess_and_predict(roi)
        if info is None:
            print("No valid digit detected. Try larger digit / better lighting.")
            current_processed = None
        else:
            current_processed = info["img28"]
            print("Mode:", info["mode"])
            print("Predicted:", info["pred"], " Confidence:", f"{info['conf']:.3f}")
            print("Top3:", info["top3"])
            # show model input enlarged
            debug = (current_processed * 255).astype(np.uint8)
            debugbig = cv2.resize(debug, (280,280), interpolation=cv2.INTER_NEAREST)
            cv2.imshow("Model sees (28x28)", debugbig)

    if key >= ord('0') and key <= ord('9'):
        if current_processed is None:
            print("No processed image to save. Press 's' first.")
        else:
            label = chr(key)
            outdir = os.path.join(SAVE_DIR, label)
            os.makedirs(outdir, exist_ok=True)
            count = len(os.listdir(outdir))
            fname = os.path.join(outdir, f"{count}.png")
            cv2.imwrite(fname, (current_processed * 255).astype(np.uint8))
            print("Saved:", fname)

    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import os

# Load model
model = load_model("my_model.keras")
print(f"Model loaded. Input shape: {model.input_shape}")

cap = cv2.VideoCapture(0)

print("Press 's' to scan & predict, 'q' to quit.")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Draw region of interest (ROI)
    x, y, w, h = 150, 100, 300, 300
    cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

    # Show live video feed
    cv2.imshow("Digit Recognition", frame)

    key = cv2.waitKey(1) & 0xFF

    # Press 's' to capture and predict
    if key == ord('s'):
        roi = frame[y:y+h, x:x+w]
        gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)

        # Invert colors if background is dark and digit is light
        _, roi_thresh = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY_INV)

        # If your model trained on black digits over white background, invert again:
        roi_final = cv2.bitwise_not(roi_thresh)

        # Resize properly without distortion
        roi_resized = cv2.resize(roi_final, (28, 28), interpolation=cv2.INTER_AREA)

        # Normalize
        roi_input = roi_resized.astype("float32") / 255.0
        roi_input = roi_input.reshape(1, 28, 28, 1)

        # Predict
        prediction = model.predict(roi_input)
        predicted_class = np.argmax(prediction)
        confidence = np.max(prediction)

        print(f"Predicted: {predicted_class} | Confidence: {confidence:.3f}")
        cv2.imshow("Processed ROI", roi_resized)

    # Press 'q' to quit
    elif key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import os

model = load_model("my_model.keras")

os.makedirs("collected_samples", exist_ok=True)

cap = cv2.VideoCapture(0)
print("Press 's' to capture & predict | Press 'q' to quit")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Draw ROI rectangle
    x, y, w, h = 150, 100, 300, 300
    cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

    # Show live video feed
    cv2.imshow("Digit Scanner", frame)

    key = cv2.waitKey(1) & 0xFF

    # When user presses 's', capture frame and predict
    if key == ord('s'):
        roi = frame[y:y+h, x:x+w]
        gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)

        # Invert for MNIST (black digit on white background)
        _, roi_thresh = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY_INV)

        # Resize and normalize
        roi_resized = cv2.resize(roi_thresh, (28, 28), interpolation=cv2.INTER_AREA)
        roi_input = roi_resized.astype("float32") / 255.0
        roi_input = roi_input.reshape(1, 28, 28, 1)

        # Predict
        prediction = model.predict(roi_input)
        predicted_class = np.argmax(prediction)
        confidence = prediction[0][predicted_class]

        # Show prediction on frame
        result_text = f"Predicted: {predicted_class} ({confidence:.2f})"
        cv2.putText(frame, result_text, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        cv2.imshow("Prediction", frame)

    # Quit the app
    elif key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
