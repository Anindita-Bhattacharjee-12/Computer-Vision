```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

model=load_model("my_model.keras")
cap=cv2.VideoCapture(0)

while True:
    ret,frame=cap.read()
    if not ret:
        break

    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    x,y,w,h=150,100,300,300
    cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
    roi=gray[y:y+h,x:x+w]

    
    _, roi_thresh=cv2.threshold(roi,128,255,cv2.THRESH_BINARY_INV)

    
    roi_resized=cv2.resize(roi_thresh,(24,24),interpolation=cv2.INTER_AREA)

    roi_padded=cv2.copyMakeBorder(roi_resized,2,2,2,2,cv2.BORDER_CONSTANT,value=0)

    roi_input=roi_padded.astype("float32")/255.0
    roi_input=roi_input.reshape(1,28,28,1)

    prediction=model.predict(roi_input)
    predicted_class=np.argmax(prediction)

    cv2.putText(frame,f"Prediction: {predicted_class}",(50,50),
                cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)

    cv2.imshow("Digit Recognition",frame)

    if cv2.waitKey(1) & 0xFF== ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

```
